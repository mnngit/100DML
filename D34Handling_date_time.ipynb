# Working with dates

```python
# converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1.Extract year

```python
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

```python
date['date_month_no'] = date['date'].dt.month
```

```python
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
```

```python
date.head()
```

# extract days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# day of the week
date['date_of_week'] = date['date'].dt.dayofweek
date.head()
```

```python
date['date_of_week'] = date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['date_is_weekend'] = np.where(date['date_of_week'].isin(['Saturday','Sunday']), 1, 0)
```

```python
date.head()
```

# Extract week of the year

```python
date['date_week'] = date['date'].dt.week
date.head()
```

# Extract Quarter

```python
date['yearly_quarter'] = date['date'].dt.quarter
```

```python
date.head()
```

# Extract Semester

```python
date['Semester'] = np.where(date['yearly_quarter'].isin([1,2]),1,2)
```

```python
date.head()
```

# Extract time elapsed between dates

```python
import datetime

today = datetime.datetime.today()
```

```python
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
np.round((today - date['date']) / np.timedelta64(1, 'M'), 0)
```

```python
# converting to datetime datatype
time['date'] = pd.to_datetime(time['date'])
```

```python
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['minute'] = time['date'].dt.minute
time['second'] = time['date'].dt.second

time.head()
```

# Extract Time Part

```python
time['time'] = time['date'].dt.time
time.head()
```

```python
# time difference
today - time['date']
```

```python
# in second
(today - time['date']) / np.timedelta64(1, 's')
```

```python
# in minutes
(today - time['date']) / np.timedelta64(1, 'm')
```

```python
# in hours
(today - time['date']) / np.timedelta64(1, 'h')
```
